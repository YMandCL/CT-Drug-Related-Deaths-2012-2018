We need to count specific drug deaths by converting "Y" in each column

```python
def drug(value, column):
    if(value == 'Y'):
        return str(column).upper()
    else:
        return value

for column in od.loc[:, 'Heroin':'AnyOpioid']:
    if(column != 'Other'):
        od[column] = od[column].map(lambda value: drug(value, column))
        
od.loc[:, 'Heroin':'AnyOpioid'].head()
```

```python
df1 = pd.DataFrame({'col': ['FENTANYL']})
df1
```

```python
#Create a new column and append the values in the drug columns to track multiple-use
od['drugs_used'] = od.loc[:, 'Heroin':'AnyOpioid'].apply(lambda value: ', '.join(value[value.notnull()]), axis = 1)
od['drugs_used'].value_counts()
```

```python
drugs_used = od['drugs_used'].value_counts()
drugs_used.loc[drugs_used > 100]
```

```python
od['drugs_count'] = od['drugs_used'].apply(lambda drugs: len(drugs.split(', ')))
od.head()
```

```python
drugs_invovled_counts = od.groupby('drugs_count')['drugs_count'].count()
```

```python
drugs_invovled_counts.plot(kind='bar', rot=0, figsize=(10,7), colormap='viridis');
plt.xlabel("Count of drugs involved")
plt.ylabel("Number of deaths")
```

```python
# Deaths caused when 1 drug was involved
import seaborn as sns
num_of_drugs_df = od.groupby(['drugs_count', 'drugs_used']).size().reset_index(name='count') 
one_drug_df = num_of_drugs_df[(num_of_drugs_df['drugs_count'] == 1) & (num_of_drugs_df['count'] >= 50)].sort_values('count', ascending=False).head()
one_drug_df.plot(x="drugs_used", y="count", kind="barh", figsize=(8,6), colormap='viridis')
plt.xlabel("Number of deaths where only 1 Drug was involved")
plt.ylabel("Drugs Used")
```

```python
od['Fent'] = od['COD'].str.contains('fent', case=False).astype(int)
```

```python
od['Fent']
```

```python
# Add more Date columns
date_new = pd.to_datetime(od['Date'])
od.insert(loc=2, column='new_date', value=date_new)
od.insert(loc=3, column='Year', value=od['new_date'].dt.year)
od.insert(loc=4, column='Month', value=od['new_date'].dt.month)
months={1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
od['Month'] = od['Month'].apply(lambda x: months[x])

od.insert(loc=5,column='day_of_week',value=od['new_date'].dt.dayofweek)
days = {0:'Mon',1:'Tue',2:'Wed',3:'Thur',4:'Fri',5:'Sat',6:'Sun'}
od['day_of_week'] = od['day_of_week'].apply(lambda x: days[x])
```

```python
fent_death = od['Fent'].value_counts()
```

```python
# Pivot Table
fent_death = od.pivot_table('Age', index='Year', columns = 'Fent', aggfunc='count');
fent_death
```

```python
fent_death.plot(kind='line', figsize = (10,7), fontsize = 13, color = ['white', 'gold'], linewidth=5)
plt.ylabel('Fentanyl Related Deaths', fontsize = 14)
plt.title('307% Increase in Fentanyl Related Deaths Since 2015', fontsize=14)
plt.show()
```
